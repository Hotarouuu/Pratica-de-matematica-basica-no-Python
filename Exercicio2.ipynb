### Exercício 8: Equação Quadrática com Discriminante

Para a equação quadrática ax^2 + bx + c = 0, calcule o discriminante e determine o número de raízes reais.

O discriminante (mais conhecido como delta) classifica as raízes da equação em tres casos:

delta > 0 -> duas raízes reais diferentes


delta = 0 -> duas raízes reais iguais



delta < 0

```python
a = 1
b = -5
c = 6

delta = b**2 - 4*a*c

x1 = (-b+math.sqrt((b**2)-(4*(a*c))))/(2*a)
x2 = (-b-math.sqrt((b**2)-(4*(a*c))))/(2*a)

if delta > 0:
    print(f'A equação possui duas raizes diferentes: X1 = {x1} e X2 = {x2} ')
elif delta == 0:
    print(f'A equação possui duas raizes iguais: X1 = {x1} e X2 = {x2} ')
elif delta < 0:
    print('A raiz não possui solução.')
```
